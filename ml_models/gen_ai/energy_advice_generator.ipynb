```python
# energy_advice_generator.ipynb

# Import necessary libraries
import pandas as pd
import openai  # Assuming OpenAI's GPT model is used
import numpy as np
import random

# Set up OpenAI API key
openai.api_key = 'your_openai_api_key'

# Load user data (assume a similar dataset structure)
df = pd.read_csv('user_energy_data.csv')

# Sample user data for generating advice
sample_user = df.iloc[random.randint(0, len(df) - 1)]
user_profile = {
    "user_id": sample_user['user_id'],
    "average_usage": sample_user['average_usage'],
    "peak_hours_usage": sample_user['peak_hours_usage'],
    "off_peak_hours_usage": sample_user['off_peak_hours_usage'],
    "previous_month_usage": sample_user['previous_month_usage']
}

# Construct the prompt for the generative AI model
prompt = f"""
User Profile:
- User ID: {user_profile['user_id']}
- Average Daily Usage: {user_profile['average_usage']} kWh
- Peak Hours Usage: {user_profile['peak_hours_usage']} kWh
- Off-Peak Hours Usage: {user_profile['off_peak_hours_usage']} kWh
- Previous Month Usage: {user_profile['previous_month_usage']} kWh

Based on the above user profile, provide personalized advice on optimizing energy consumption:
"""

# Generate advice using OpenAI GPT model
response = openai.Completion.create(
    engine="davinci",
    prompt=prompt,
    max_tokens=150,
    temperature=0.7
)

# Print the generated advice
print(f"Generated Advice for User {user_profile['user_id']}:\n")
print(response.choices[0].text.strip())

# Save the generated advice
advice_output = {
    "user_id": user_profile['user_id'],
    "advice": response.choices[0].text.strip()
}
output_df = pd.DataFrame([advice_output])
output_df.to_csv(f"advice_for_user_{user_profile['user_id']}.csv", index=False)
```
